```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
plt.style.use('fivethirtyeight')
```

```python
#Countries reader
def filter_country(country):
    countries = pd.read_csv(
         'dailypricespercountry.csv',
         parse_dates=['Date'])
    country_data=countries.loc[(countries['Country']==country)  ]
    country_data['Date'] = pd.to_datetime(country_data['Date'])
    country_data.set_index('Date', inplace=True)
    filtered_country_data=country_data.loc['2019-01-01':'2023-12-31']
    return filtered_country_data
```

```python

def create_lstm_model(future_name, country_name):
    # Read future data
    future_data = pd.read_csv('futures/' + future_name + '.csv', parse_dates=['Date'])
    future_data['Date'] = pd.to_datetime(future_data['Date'])
    future_data.set_index('Date', inplace=True)

    # Sort datetime index
    future_data.sort_index(inplace=True)
    
    # Slice future data for the desired period
    future_data = future_data.loc['2019-01-01':'2023-12-31']

    # Filter country data
    country_data = filter_country(country_name)

    # Merge future and country data
    combined_data = pd.merge(future_data, country_data, on='Date', how='inner')

    # Extract features and target
    features = combined_data[['Price', 'Price (EUR/MWhe) ']]
    target = combined_data['Price (EUR/MWhe) ']

    # Scale the data
    scaler_features = MinMaxScaler(feature_range=(0, 1))
    scaler_target = MinMaxScaler(feature_range=(0, 1))

    scaled_features = scaler_features.fit_transform(features)
    scaled_target = scaler_target.fit_transform(target.values.reshape(-1, 1))

    # Define window size
    window_size = 60

    # Prepare data for LSTM
    X, y = [], []
    for i in range(window_size, len(combined_data)):
        X.append(scaled_features[i-window_size:i, :])
        y.append(scaled_target[i, 0])

    # Convert to numpy arrays
    X, y = np.array(X), np.array(y)

    # Train-test split (you may want to adjust this)
    split_ratio = 0.8
    split_index = int(len(X) * split_ratio)

    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]

    # Build LSTM model
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
        LSTM(50, return_sequences=False),
        Dense(25),
        Dense(1)
    ])

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(X_train, y_train, batch_size=64, epochs=10, validation_data=(X_test, y_test), verbose=1)
  
    

    # Predict electricity prices
    predictions = model.predict(X_test)
    predictions = scaler_target.inverse_transform(predictions).flatten()

     #Get the root mean squared error(RMSE)
    rmse=np.sqrt(np.mean((predictions-y_test)**2))
    print("RMSE:", rmse)
    
    plt.figure(figsize=(12, 6))
    plt.plot(combined_data.index[:split_index], scaler_target.inverse_transform(y_train.reshape(-1, 1)), label='Training Set (Actual)', color='blue')

    # Slice the index for the test set
    test_index = combined_data.index[split_index:split_index+len(y_test)]

    plt.plot(test_index, scaler_target.inverse_transform(y_test.reshape(-1, 1)), label='Test Set (Actual)', color='red')
    plt.plot(test_index, predictions, label='Predicted Electricity Price', color='green')
    if future_name=='Brent':
        plt.title('Greece Predicted vs Actual Electricity Prices - Brent Model')
    elif future_name=='TTF':
        plt.title('Greece Predicted vs Actual Electricity Prices - TTF Model')
    elif future_name=='EUAs':
        plt.title('Greece Predicted vs Actual Electricity Prices - EUAs Model')
    plt.xlabel('Date')
    plt.ylabel('Price (EUR/MWhe)')
    plt.legend()
    plt.show()


# Example usage
create_lstm_model('TTF', 'Greece')
create_lstm_model('EUAs', 'Greece')
create_lstm_model('Brent', 'Greece')



```
