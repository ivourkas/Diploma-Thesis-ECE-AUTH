```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
def countries_plot(country):
    countries = pd.read_csv(
         'dailypricespercountry.csv',
         parse_dates=['Date'])
    country_data=countries.loc[(countries['Country']==country)  ]
    country_data['Date'] = pd.to_datetime(country_data['Date'])
    country_data.set_index('Date', inplace=True)
    filtered_country_data=country_data.loc['2019-01-01':'2023-12-31']
    plt.figure(figsize=(15, 6))
    plt.plot(filtered_country_data.index, filtered_country_data['Price (EUR/MWhe) '], label='Electricity Price')
    plt.xlabel('Date')
    plt.ylabel('Price (EUR/MWhe) ')
    plt.title( country + ' Electricity Price')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
countries_plot('Austria')
countries_plot('Belgium')
countries_plot('Czechia')
countries_plot('Denmark')
countries_plot('Finland')
countries_plot('France')
countries_plot('Germany')
countries_plot('Greece')
countries_plot('Hungary')
countries_plot('Bulgaria')
countries_plot('Italy')
countries_plot('Netherlands')
countries_plot('Norway')
countries_plot('Poland')
countries_plot('Portugal')
countries_plot('Romania')
countries_plot('Serbia')
countries_plot('Spain')
countries_plot('Switzerland')
countries_plot('Sweden')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate

def countries_statistics(start_date, end_date):
    # Read data
    countries = pd.read_csv('dailypricespercountry.csv', parse_dates=['Date'])
    
    # Define countries and time period
    country_list = ['Austria', 'Belgium', 'Czechia', 'Denmark', 'Finland', 'France', 
                    'Germany', 'Greece', 'Hungary', 'Bulgaria', 'Italy', 'Netherlands', 
                    'Norway', 'Poland', 'Portugal', 'Romania', 'Serbia', 'Spain', 
                    'Switzerland', 'Sweden']
    
    # Initialize empty DataFrame to store statistics
    statistics_df = pd.DataFrame(columns=['Country', 'N', 'Range', 'Minimum', 'Maximum', 'Mean', 'Std. Deviation', 'Variance'])
    
    # Compute statistics for each country
    for country in country_list:
        country_data = countries.loc[(countries['Country'] == country) & 
                                     (countries['Date'] >= start_date) & 
                                     (countries['Date'] <= end_date)]
        
        # Compute descriptive statistics
        N = len(country_data)
        price_range = country_data['Price (EUR/MWhe) '].max() - country_data['Price (EUR/MWhe) '].min()
        minimum = country_data['Price (EUR/MWhe) '].min()
        maximum = country_data['Price (EUR/MWhe) '].max()
        mean = country_data['Price (EUR/MWhe) '].mean()
        std_deviation = country_data['Price (EUR/MWhe) '].std()
        variance = country_data['Price (EUR/MWhe) '].var()
        
        # Append statistics to DataFrame
        statistics_df = pd.concat([statistics_df, pd.DataFrame({'Country': country, 'N': N, 'Range': price_range,
                                              'Minimum': minimum, 'Maximum': maximum, 'Mean': mean,
                                              'Std. Deviation': std_deviation, 'Variance': variance}, 
                                             index=[0])], ignore_index=True)
    
    # Convert DataFrame to tabular format
    table = tabulate(statistics_df, headers='keys', tablefmt='grid')
    
    # Determine title based on start and end dates
    if start_date == '2020-03-11' and end_date == '2021-08-23':
        title = 'COVID-19 period statistics'
    elif start_date == '2022-02-24' and end_date == '2023-02-24':
        title = 'Ukraine War period statistics'
    elif start_date == '2023-10-07' and end_date == '2023-12-31':
        title = 'Israel War period statistics'
    else:
        title = 'Custom period statistics'
    
    # Plot the table as an image
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.axis('off')
    ax.set_title(title)  # Set the title
    ax.table(cellText=statistics_df.values,
             colLabels=statistics_df.columns,
             cellLoc='center',
             loc='center')
    plt.savefig('statistics_table.png', bbox_inches='tight', pad_inches=0.1)
    plt.show()

# Example usage:
countries_statistics('2020-03-11', '2021-08-23')
countries_statistics('2022-02-24', '2023-02-24')
countries_statistics('2023-10-07','2023-12-31')
```
